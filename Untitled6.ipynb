```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Plot settings
plt.rcParams["figure.figsize"] = (9, 5)
plt.rcParams["axes.grid"] = True
pd.set_option("display.max_columns", 100)

print("Libraries imported ✅")
```

```python
from google.colab import files

print("Please select your CSV file in the dialog…")
uploaded = files.upload()  # shows a file picker
file_name = next(iter(uploaded))  # first uploaded file
print(f"Loaded file: {file_name}")

# Read CSV (auto-detect common encodings)
read_attempts = ["utf-8", "latin-1", "cp1252"]
for enc in read_attempts:
    try:
        df = pd.read_csv(file_name, encoding=enc)
        print(f"Read with encoding: {enc}")
        break
    except Exception as e:
        last_err = e
else:
    raise last_err

print("Shape:", df.shape)
display(df.head())
```

```python
# Basic info
print("Data types:")
print(df.dtypes)

# Missing values
na_counts = df.isna().sum().sort_values(ascending=False)
print("\nMissing values summary:")
display(na_counts[na_counts > 0])

# Try to auto-detect datetime columns
def detect_datetime(series: pd.Series) -> bool:
    if series.dtype == "object":
        try:
            pd.to_datetime(series.dropna().sample(min(100, series.dropna().shape[0])), errors="raise")
            return True
        except Exception:
            return False
    return pd.api.types.is_datetime64_any_dtype(series)

dt_cols = [c for c in df.columns if detect_datetime(df[c])]
for c in dt_cols:
    df[c] = pd.to_datetime(df[c], errors="coerce")

print("\nDetected datetime columns:", dt_cols if dt_cols else "None")

# Separate numeric & categorical columns
num_cols = df.select_dtypes(include=[np.number]).columns.tolist()
cat_cols = df.select_dtypes(include=["object", "category", "bool"]).columns.tolist()

print("\nNumeric columns:", num_cols if num_cols else "None")
print("Categorical columns:", cat_cols if cat_cols else "None")

# Summary stats
print("\nSummary statistics (numeric):")
display(df[num_cols].describe().T if num_cols else "No numeric columns")
print("\nSummary statistics (categorical):")
display(df[cat_cols].describe().T if cat_cols else "No categorical columns")

```

```python
MAX_PLOTS = 8  # limit to avoid too many figures

if num_cols:
    cols_to_plot = num_cols[:MAX_PLOTS]
    for col in cols_to_plot:
        fig, ax = plt.subplots()
        sns.histplot(df[col].dropna(), kde=True, ax=ax)
        ax.set_title(f"Distribution of {col}")
        ax.set_xlabel(col)
        plt.show()
else:
    print("No numeric columns to plot.")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

MAX_CATS = 12  # show up to 12 categories per feature

if cat_cols:
    cols_to_plot = cat_cols[:6]  # limit number of categorical columns to visualize
    for col in cols_to_plot:
        vc = df[col].astype("category").value_counts().head(MAX_CATS)

        # --- Bar Chart with different colors ---
        fig, ax = plt.subplots()
        vc.plot(kind="bar", ax=ax, color=sns.color_palette("Set2", len(vc)))
        ax.set_title(f"Category Counts: {col} (Bar)")
        ax.set_xlabel(col)
        ax.set_ylabel("Count")
        plt.xticks(rotation=45, ha="right")
        plt.tight_layout()
        plt.show()

        # --- Pie Chart with different colors + fixed title position ---
        fig, ax = plt.subplots()
        vc.plot(
            kind="pie",
            autopct=lambda p: '{:.1f}%'.format(p) if p > 3 else '',  # show only if >3%
            startangle=90,
            colors=sns.color_palette("tab10", len(vc)),
            ax=ax
        )
        ax.set_ylabel("")  # remove y-label for cleaner look
        ax.set_title(f"Category Distribution: {col} (Pie)", y=1.08)  # push title above pie
        plt.tight_layout()
        plt.show()
else:
    print("No categorical columns to plot.")
```

```python
# Try to auto-detect a likely target column by name
possible_targets = ["target","label","class","churn","default","Survived","Outcome","y"]
found_targets = [c for c in df.columns if c in possible_targets]

TARGET = found_targets[0] if found_targets else None  # ← set manually to a column name if needed

print("Detected TARGET:", TARGET)

if TARGET is None:
    print("No obvious target detected. If you want target-based analysis, set TARGET = 'your_column_name' and rerun this cell.")
else:
    if df[TARGET].dtype in ["object", "bool", "category"] or df[TARGET].nunique() <= 10:
        # Classification-like comparisons
        print(f"Target '{TARGET}' looks categorical/classification-like.")
        print("Class balance:")
        display(df[TARGET].value_counts(dropna=False))

        # Numeric features by class (boxplots)
        for num in num_cols[:6]:
            fig, ax = plt.subplots()
            sns.boxplot(x=df[TARGET], y=df[num], ax=ax)
            ax.set_title(f"{num} by {TARGET}")
            ax.set_xlabel(TARGET)
            ax.set_ylabel(num)
            plt.xticks(rotation=0)
            plt.tight_layout()
            plt.show()

        # Categorical features vs target (stacked percentages)
        for cat in cat_cols[:4]:
            ctab = pd.crosstab(df[cat], df[TARGET], normalize="index") * 100
            fig, ax = plt.subplots()
            ctab.plot(kind="bar", stacked=True, ax=ax)
            ax.set_title(f"{cat} vs {TARGET} (%)")
            ax.set_ylabel("Row %")
            plt.xticks(rotation=30, ha="right")
            plt.tight_layout()
            plt.show()
    else:
        # Regression-like comparisons
        print(f"Target '{TARGET}' looks numeric/regression-like.")
        # Scatter plots vs top numeric features with Pearson r
        top_features = [c for c in num_cols if c != TARGET][:6]
        for num in top_features:
            valid = df[[num, TARGET]].dropna()
            if valid.shape[0] < 5:
                continue
            r, p = stats.pearsonr(valid[num], valid[TARGET])
            fig, ax = plt.subplots()
            ax.scatter(valid[num], valid[TARGET], alpha=0.6)
            ax.set_title(f"{TARGET} vs {num} (r={r:.2f}, p={p:.3g})")
            ax.set_xlabel(num)
            ax.set_ylabel(TARGET)
            plt.tight_layout()
            plt.show()
```

```python
# Quick preview helpers; they DO NOT modify df unless you set apply_changes=True
apply_changes = False  # set to True if you want to apply

df_clean = df.copy()

# Simple numeric imputation with median
if num_cols:
    medians = df[num_cols].median()
    if apply_changes:
        df_clean[num_cols] = df_clean[num_cols].fillna(medians)
    print("Numeric columns median (for potential imputation):")
    display(medians)

# Simple categorical imputation with mode
if cat_cols:
    modes = df[cat_cols].mode(dropna=True).iloc[0]
    if apply_changes:
        for c in cat_cols:
            df_clean[c] = df_clean[c].fillna(modes[c])
    print("Categorical columns mode (for potential imputation):")
    display(modes)

if apply_changes:
    print("Applied simple imputations. New shape (unchanged rows/cols):", df_clean.shape)
else:
    print("No changes applied. Set apply_changes=True to create df_clean with imputations.")
```

```python
import pandas as pd
import numpy as np
import plotly.express as px
from google.colab import files

# Upload your CSV
uploaded = files.upload()
file_name = list(uploaded.keys())[0]
df = pd.read_csv("Social media.csv")

print("Shape:", df.shape)
df.head()
```

```python
import plotly.figure_factory as ff

num_cols = df.select_dtypes(include=np.number).columns.tolist()

if len(num_cols) >= 2:
    corr = df[num_cols].corr()
    fig = ff.create_annotated_heatmap(
        z=corr.values,
        x=list(corr.columns),
        y=list(corr.index),
        colorscale="Viridis",
        showscale=True
    )
    fig.update_layout(title="Correlation Heatmap (Interactive)")
    fig.show()
```

```python
for col in num_cols:
    fig = px.histogram(df, x=col, nbins=30, title=f"Distribution of {col}", color_discrete_sequence=["skyblue"])
    fig.update_traces(marker_line_width=1, marker_line_color="black")
    fig.show()
```
