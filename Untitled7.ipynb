```python
%%writefile app.py
import streamlit as st
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px

st.set_page_config(page_title="Interactive Dashboard", layout="wide")

st.title("📊 Interactive Data Dashboard")

# File uploader
uploaded_file = st.file_uploader("Upload your CSV file", type=["csv"])
if uploaded_file:
    df = pd.read_csv(uploaded_file)
    st.success(f"Dataset Loaded! Shape: {df.shape}")

    st.subheader("🔎 Data Preview")
    st.dataframe(df.head())

    # Detect numeric & categorical columns
    num_cols = df.select_dtypes(include="number").columns.tolist()
    cat_cols = df.select_dtypes(exclude="number").columns.tolist()

    # Summary Stats
    st.subheader("📌 Summary Statistics")
    st.write(df.describe(include="all"))

    # Missing values
    st.subheader("⚠️ Missing Values")
    st.write(df.isna().sum()[df.isna().sum() > 0])

    # Correlation Heatmap
    if len(num_cols) >= 2:
        st.subheader("📈 Correlation Heatmap")
        corr = df[num_cols].corr()
        fig = px.imshow(corr, text_auto=True, color_continuous_scale="Viridis")
        st.plotly_chart(fig, use_container_width=True)

    # Histograms
    st.subheader("📊 Histograms")
    col_choice = st.selectbox("Select Numeric Column", num_cols)
    if col_choice:
        fig = px.histogram(df, x=col_choice, nbins=30, title=f"Distribution of {col_choice}")
        st.plotly_chart(fig, use_container_width=True)

    # Boxplots
    if num_cols and cat_cols:
        st.subheader("📦 Boxplots")
        num_choice = st.selectbox("Select Numeric Column", num_cols, key="num_box")
        cat_choice = st.selectbox("Select Categorical Column", cat_cols, key="cat_box")
        if num_choice and cat_choice:
            fig = px.box(df, x=cat_choice, y=num_choice, color=cat_choice, title=f"{num_choice} across {cat_choice}")
            st.plotly_chart(fig, use_container_width=True)

    # Bar + Pie Charts
    st.subheader("🍩 Categorical Distribution")
    cat_choice2 = st.selectbox("Select Categorical Column", cat_cols, key="cat_plot")
    if cat_choice2:
        vc = df[cat_choice2].value_counts().reset_index()
        vc.columns = [cat_choice2, "Count"]

        col1, col2 = st.columns(2)

        with col1:
            fig = px.bar(vc, x=cat_choice2, y="Count", color=cat_choice2, title=f"Counts of {cat_choice2}")
            st.plotly_chart(fig, use_container_width=True)

        with col2:
            fig = px.pie(vc, names=cat_choice2, values="Count", title=f"Pie Chart of {cat_choice2}")
            st.plotly_chart(fig, use_container_width=True)
else:
    st.warning("Please upload a CSV file to get started.")
```

```python
# Step 1: Install required libraries
!pip install plotly pandas ipywidgets

# Step 2: Import libraries
import pandas as pd
import plotly.express as px
import ipywidgets as widgets
from IPython.display import display

# Step 3: Load your dataset (replace with your uploaded filename)
df = pd.read_csv("Social media.csv")

# Step 4: Separate numeric and categorical columns
num_cols = df.select_dtypes(include='number').columns.tolist()
cat_cols = df.select_dtypes(exclude='number').columns.tolist()

# Step 5: Create dropdowns
num_dropdown = widgets.Dropdown(options=num_cols, description="Numeric:")
cat_dropdown = widgets.Dropdown(options=cat_cols, description="Category:")

# Step 6: Define update function
def update_dashboard(numeric, category):
    # Histogram
    if numeric:
        fig = px.histogram(df, x=numeric, nbins=20, title=f"Histogram of {numeric}")
        fig.show()

        # Boxplot
        if category:
            fig = px.box(df, x=category, y=numeric, color=category, title=f"Boxplot of {numeric} by {category}")
            fig.show()

    # Bar chart + Pie chart
    if category:
        fig = px.bar(df, x=category, color=category, title=f"Bar Chart of {category}")
        fig.show()

        fig = px.pie(df, names=category, title=f"Pie Chart of {category}")
        fig.show()

    # Correlation Heatmap
    if num_cols:
        corr = df[num_cols].corr()
        fig = px.imshow(corr, text_auto=True, aspect="auto", title="Correlation Heatmap")
        fig.show()

# Step 7: Interactive output
widgets.interact(update_dashboard, numeric=num_dropdown, category=cat_dropdown);
```
